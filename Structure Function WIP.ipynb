```python
def structure_function(y,tau, pairs = 50):
    N = []
    s = 0 
    ivals = range(len(y))
    for k in range(pairs):
        i = np.random.choice(ivals)
        j = tau + i
        if i < j < len(y): 
            s += (y[i]-y[j])**2
            N.append(1)
    return np.sqrt((1/sum(N))*s)
```

```python
x = ls2_ch1_0x
y = ls2_ch1_0y

plt.figure(figsize=(12,8))

plt.title('Raw Data')
plt.plot(x,y,color='darkslateblue')
plt.xlabel('Time Samples')
plt.ylabel('Voltage [V]')
plt.show()

```

```python
taus = np.linspace(0.0001,100,1000)
sf = [structure_function(y,int(tau)) for tau in taus]

freqs = [60,120,180,240]
colors = ['blue','green','orange','black']

plt.figure(figsize=(12,8))
plt.loglog(taus,sf,'r')
for i in range(4):
    plt.axvline(600/freqs[i],color=colors[i],label=str(freqs[i])+'Hz')
plt.xlabel(r'log($\tau$)')
plt.ylabel(r'log(Sf($\tau$))')
plt.legend()
plt.title('Structure Function')
plt.show()
```

```python
x_lin = np.linspace(0,1,1000)
y_cos = np.cos(2*pi*60*x_lin)
plt.plot(x_lin,y_cos)
plt.show()

taus = np.linspace(0,100,100)
sf = [structure_function(y_cos,int(tau),pairs=100) for tau in taus]
```

## Spectrograms

```python
fs = 600
y1 = ls2_ch1_0y
y2 = ls2_ch2_0y
y3 = ls2_ch3_0y
y4 = ls2_ch4_0y

f1, t1, Sxx1 = signal.spectrogram(y1, fs)
f2, t2, Sxx2 = signal.spectrogram(y2, fs)
f3, t3, Sxx3 = signal.spectrogram(y3, fs)
f4, t4, Sxx4 = signal.spectrogram(y4, fs)
```
