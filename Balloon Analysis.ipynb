```python
%pylab inline
from scipy import signal
```

## Balloon Data

```python
"""
Channel 1
"""
data1 = np.genfromtxt("LS1CH1_5110.0-5180.0sec.txt", autostrip = True)
ls1_ch1_0x, ls1_ch1_0y = data1[:,0],data1[:,1]
data2 = np.genfromtxt("LS1CH1_5330.0-5400.0sec.txt", autostrip = True)
ls1_ch1_1x, ls1_ch1_1y = data2[:,0],data2[:,1]

"""
Channel 2
"""
data3 = np.genfromtxt("LS1CH2_5110.0-5180.0sec.txt", autostrip = True)
ls1_ch2_0x, ls1_ch2_0y = data3[:,0],data3[:,1]
data4 = np.genfromtxt("LS1CH2_5330.0-5400.0sec.txt", autostrip = True)
ls1_ch2_1x, ls1_ch2_1y = data4[:,0],data4[:,1]

"""
Channel 3
"""
data5 = np.genfromtxt("LS1CH3_5110.0-5180.0sec.txt", autostrip = True)
ls1_ch3_0x, ls1_ch3_0y = data5[:,0],data5[:,1]
data6 = np.genfromtxt("LS1CH3_5330.0-5400.0sec.txt", autostrip = True)
ls1_ch3_1x, ls1_ch3_1y = data6[:,0],data6[:,1]

"""
Channel 4
"""
data7 = np.genfromtxt("LS1CH4_5110.0-5180.0sec.txt", autostrip = True)
ls1_ch4_0x, ls1_ch4_0y = data7[:,0],data7[:,1]
data8 = np.genfromtxt("LS1CH4_5330.0-5400.0sec.txt", autostrip = True)
ls1_ch4_1x, ls1_ch4_1y = data8[:,0],data8[:,1]

```

```python
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.plot(ls1_ch1_0x, ls1_ch1_0y)
plt.title('Channel 1')

plt.subplot(2,2,2)
plt.plot(ls1_ch2_0x, ls1_ch2_0y)
plt.title('Channel 2')

plt.subplot(2,2,3)
plt.plot(ls1_ch3_0x, ls1_ch3_0y)
plt.title('Channel 3')

plt.subplot(2,2,4)
plt.plot(ls1_ch4_0x, ls1_ch4_0y)
plt.title('Channel 4')

plt.tight_layout()
plt.show()
```

## Spectrograms

```python
fs = 600
y1 = ls1_ch1_0y
y2 = ls1_ch2_0y
y3 = ls1_ch3_0y
y4 = ls1_ch4_0y

f1, t1, Sxx1 = signal.spectrogram(y1, fs)
f2, t2, Sxx2 = signal.spectrogram(y2, fs)
f3, t3, Sxx3 = signal.spectrogram(y3, fs)
f4, t4, Sxx4 = signal.spectrogram(y4, fs)
```

```python
plt.figure(figsize=(15,10))
plt.suptitle('Balloon Array')

plt.subplot(2,2,1)
plt.pcolormesh(t1, f1, Sxx1,cmap='jet')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim(0,100)
plt.title('Channel 1')

plt.subplot(222)
plt.pcolormesh(t2, f2, Sxx2,cmap='jet')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim(0,100)
plt.title('Channel 2')

plt.subplot(223)
plt.pcolormesh(t3, f3, Sxx3,cmap='jet')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim(0,100)
plt.title('Channel 3')

plt.subplot(224)
plt.pcolormesh(t4, f4, Sxx4,cmap='jet')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.ylim(0,100)
plt.title('Channel 4')

plt.tight_layout()
plt.show()
```
