```python
"""
Functions 
"""

def find_median_4(x,y,z,k,vector = False):
    med = np.dstack((x,y,z,k))
    median_arr = np.copy(x)
    i = 0
    
    if vector == False:
        while i<len(x):
            for j in range(len(x[1])):
                median_arr[i,j] = np.median(med[i,j])
            i+=1
    else:
        while i<len(x):
            median_arr[i] = np.median(med[0][i])
            i+=1
        
    return median_arr 


def normalize(x,y,deg=2,sec=30):
    x = x[:sec*600]
    y = y[:sec*600]
    x = x-x[0]
    y  = (y-y.min())/ y.max()-y.min()
    z = np.polyfit(x, y, deg)
    p = np.poly1d(z)
    xp = np.linspace(0,max(x),len(x))
    yp = p(xp)
    return x,y,xp, yp, y/yp


def sigmaclip(x,y,deg=2,start=0,sec=30,sig=3):
    x = x[start*600:sec*600]
    y = y[start*600:sec*600]
    x = x-x[0]
    y  = (y-y.min())/ y.max()-y.min()
    z = np.polyfit(x, y, deg)
    p = np.poly1d(z)
    xp = np.linspace(0,max(x),len(x))
    yp = p(xp)
    
    #xr,yr,normx,normy,divy = normalize(x,y,sec)
    xr,yr,normx,normy,divy = x,y,xp,yp,y/yp
    
    clipped_time = np.array(())
    clipped_volt = np.array(())
    clipped_ypvolt = np.array(())
    
    k = 0
    
     
    while k <= sec-5:
        chunk = divy[k*600:(k+5)*600]
        chunkyp = yp[k*600:(k+5)*600]
        
        sigclip = sigma_clip(chunk,sigma=sig,cenfunc=np.mean,stdfunc=np.std)
        
        a = normx[np.nonzero(sigclip)]+k
        b = chunk[np.nonzero(sigclip)]
        c = chunkyp[np.nonzero(sigclip)]
        
        clipped_time = np.append(clipped_time,a)
        clipped_volt = np.append(clipped_volt,b)
        clipped_ypvolt = np.append(clipped_ypvolt,c)
        k += 5    
        

  
    return (clipped_time),(clipped_volt)*clipped_ypvolt
    #return (clipped_time),(clipped_volt),(clipped_volt)*clipped_ypvolt

def point_five_sigmaclip(x,y,deg=2,sec=30):
    x = x[:sec*600]
    y = y[:sec*600]
    x = x-x[0]
    y  = (y-y.min())/ y.max()-y.min()
    z = np.polyfit(x, y, deg)
    p = np.poly1d(z)
    xp = np.linspace(0,max(x),len(x))
    yp = p(xp)
    
    xr,yr,normx,normy,divy = x,y,xp,yp,y/yp
    
    clipped_time = np.array(())
    clipped_volt = np.array(())
    clipped_ypvolt = np.array(())
    
    for i in range(len(divy)):
        if divy[i] >= 0.78:
            clipped_time = np.append(clipped_time,normx[i])
            clipped_volt = np.append(clipped_volt,divy[i])
            clipped_ypvolt = np.append(clipped_ypvolt,yp[i])

        
    
    #return (clipped_time),clipped_volt,(clipped_volt)*clipped_ypvolt
    return (clipped_time),(clipped_volt)*clipped_ypvolt
    
def running_avg(y,w=31):
    #w = window length
    
    yp = []
    
    i = 0
    while i <= len(y) - w:
        
        yp.append(np.sum(y[i:i+w])/w)
        
        i+=1
        
    
    return np.asarray(yp)

def normalize_run_avg(x,y,w):
    yp = running_avg(y,w=w)
    sl_i = int((w-1)/2)

    ynorm = y[sl_i:-sl_i]/yp
    xnorm = x[sl_i:-sl_i]

    return xnorm,ynorm   

```

```python
def data_mask_sigclip(x1,y1,x2,y2,x3,y3,x4,y4,delt=30,deg=2,sec=60):
    
    """
    First, normalize between 0 - 1
    """
    """    
    x1,x2,x3,x4 = x1[:sec*600],x2[:sec*600],x3[:sec*600],x4[:sec*600]
    y1,y2,y3,y4 = y1[:sec*600],y2[:sec*600],y3[:sec*600],y4[:sec*600]
    x1,x2,x3,x4 = x1-x1[0],x2-x2[0],x3-x3[0],x4-x4[0]
    y1,y2,y3,y4  = (y1-y1.min())/ y1.max()-y1.min(),(y2-y2.min())/ y2.max()-y2.min(),(y3-y3.min())/ y3.max()-y3.min(),(y4-y4.min())/ y4.max()-y4.min()
    z1,z2,z3,z4 = np.polyfit(x1, y1, deg),np.polyfit(x2, y2, deg),np.polyfit(x3, y3, deg),np.polyfit(x4, y4, deg)
    p1,p2,p3,p4 = np.poly1d(z1),np.poly1d(z2),np.poly1d(z3),np.poly1d(z4)
    xp1,xp2,xp3,xp4 = np.linspace(0,max(x1),len(x1)),np.linspace(0,max(x2),len(x2)),np.linspace(0,max(x3),len(x3)),np.linspace(0,max(x4),len(x4))
    yp1,yp2,yp3,yp4 = p1(xp1),p2(xp2),p3(xp3),p4(xp4)
    
    normx1,normy1 = xp1,y1/yp1
    normx2,normy2 = xp2,y2/yp2
    normx3,normy3 = xp3,y3/yp3
    normx4,normy4 = xp4,y4/yp4
    """
    x1,x2,x3,x4 = x1[:sec*600],x2[:sec*600],x3[:sec*600],x4[:sec*600]
    x1,x2,x3,x4 = x1-x1[0],x2-x2[0],x3-x3[0],x4-x4[0]
    y1,y2,y3,y4 = y1[:sec*600],y2[:sec*600],y3[:sec*600],y4[:sec*600]
    
    w = 31
    sl_i = int((w-1)/2)
    yp1,yp2,yp3,yp4 = running_avg(y1,w=w),running_avg(y2,w=w),running_avg(y3,w=w),running_avg(y4,w=w)
    xp1,xp2,xp3,xp4 = x1[sl_i:-sl_i],x2[sl_i:-sl_i],x3[sl_i:-sl_i],x4[sl_i:-sl_i]

    normx1,normy1 = xp1-xp1[0],y1[sl_i:-sl_i]/yp1
    normx2,normy2 = xp2-xp2[0],y2[sl_i:-sl_i]/yp2
    normx3,normy3 = xp3-xp3[0],y3[sl_i:-sl_i]/yp3
    normx4,normy4 = xp4-xp4[0],y4[sl_i:-sl_i]/yp4
   
    mask_y1 = []
    mask_y2 = []
    mask_y3 = []
    mask_y4 = []
    
    delta_t = delt
    i = 0
    while i <= len(normy1)-delta_t:
        v1 = normy1[i:i+delta_t]
        v2 = normy2[i:i+delta_t]
        v3 = normy3[i:i+delta_t]
        v4 = normy4[i:i+delta_t]
        
        avg_v1,avg_v2,avg_v3,avg_v4 = np.mean(v1),np.mean(v2),np.mean(v3),np.mean(v4)
        
        
        if 1.02 > avg_v1 > 0.98:
            mask_y1.extend([1 for j in range(delta_t)])
        else:
            mask_y1.extend([0 for j in range(delta_t)]) 
        
        if 1.02 > avg_v2 > 0.98:
            mask_y2.extend([1 for j in range(delta_t)])
        else:
            mask_y2.extend([0 for j in range(delta_t)]) 
       
        if 1.02 > avg_v3 > 0.98:
            mask_y3.extend([1 for j in range(delta_t)])
        else:
            mask_y3.extend([0 for j in range(delta_t)]) 
        
        if 1.02 > avg_v4 > 0.98:
            mask_y4.extend([1 for j in range(delta_t)])
        else:
            mask_y4.extend([0 for j in range(delta_t)]) 
        
        i += delta_t
        
    mask_y1,mask_y2,mask_y3,mask_y4 = np.asarray(mask_y1),np.asarray(mask_y2),np.asarray(mask_y3),np.asarray(mask_y4)
        
    #voltage = (mask_y1*normy1 + mask_y2*normy2 + mask_y3*normy3 + mask_y4*normy4) /(mask_y1 + mask_y2 + mask_y3 + mask_y4)
    voltage = (mask_y1*normy1*yp1 + mask_y2*normy2*yp2 + mask_y3*normy3*yp3 + mask_y4*normy4*yp4) /(mask_y1 + mask_y2 + mask_y3 + mask_y4)
    
    return normx1,voltage
    
```
